# Manually by using map method #

**Map method** - **Case status**

- First consider one categorical column

- Observe how many unique lables are there 

- Assume here we are taking case_status column

- case_status two unique labels 
  
    - Denied
    
    - Certified
    
- Create a dictionary by providing key:value
    
- Here key = Denied and value = 1

- Key = Certified and value = 0

- dictionary will looks : {"certified":0 , "Denied":1}

```python
visa_df["case_status"].unique()
```

```python
d1={"Certified":0 , "Denied":1}
visa_df["case_status"].map(d1)
```

```python
# After mapping in which column you want to apply the data 
# You want to create a new col or 
# you want to use old col
# it is our wish completely
visa_df["case_status"]=visa_df["case_status"].map(d1)
visa_df["case_status"]
```

```python
visa_df
```

```python
visa_df["continent"].unique()
```

```python
d2={'Asia':0, 'Africa':1, 'North America':2, 'Europe':3, 'South America':4,'Oceania':5}
visa_df["continent"]=visa_df["continent"].map(d1)
visa_df.head()
```

```python
d1={}
labels=visa_df["continent"].unique()
for i in range(len(labels)):
    d1[labels[i]]=i
d1
```

```python
visa_df["continent"]=visa_df["continent"].map(d1)
visa_df.head()
```

# Method -3 # 

**One hot encoder**

**pd.get_dummies()**

- one hot encoder means at a time only one will be ON and , Other will be OFF

- ON reperesents with 1 and OFF represents 0

- One hot encoder creates new extra  column based on labels in the columns

- For example case_status has two unique labels

- If you apply one hot encoder on case_status it creates two extra columns
 
    - case_status_Denied
    
    - case_status_Certified
    
|case_status| case_status_Denied | case_status_Certified |
------------|--------------------| ----------------------|
| case_status_Denied | 1 | 0 |
|case_status_Certified | 0 | 1 |


**Advantages**

-  One hot encoder gives independency between the variables

- It is one of the important property before apply ML models

- Variables are independent to each other

- Variables Are perpendicular to each other

- Variables are 90 degrees phase shift

- Variables are Orthogonal each other

**Disadvantages**

- Assume that a column has 100 unique labels

- If we apply one hot encoder it will creates 100 new columns along with prvious columns

- your data matrix becomes parse 

- We required more time more memory to compute the calculation

- This is called **Curse of Dimensionality**

-
