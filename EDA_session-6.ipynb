```python
# to deal with the outliers
# Drop : Not reccomonded >2% 
# Median value: outliers doesn not effect median value
# Two types of dataframes are there
# outliers_df and non_outliers_df
```

```python
Q1=np.quantile(visa_df["prevailing_wage"],0.25)
Q2=np.quantile(visa_df["prevailing_wage"],0.50)
Q3=np.quantile(visa_df["prevailing_wage"],0.75)

IQR=Q3-Q1
UB=Q3+1.5*IQR
LB=Q1-1.5*IQR

############# Outliers_df #####################
con1=visa_df["prevailing_wage"]>UB
con2=visa_df["prevailing_wage"]<LB
outliers_df=visa_df[con1 | con2]

############### non_outliers_df ###############
con11=visa_df["prevailing_wage"]<UB
con22=visa_df["prevailing_wage"]>LB
non_outliers_df=visa_df[con11 & con22]


```

# np.where

- np.where is used to fill the values based on the true condition 

- np.where will take 3 arguments 
          
     - condition
        
     - Value you want to replace if the con is true
        
     - Value you want to replace if the cond is false

**np.where(condition,True val,False val)**

```python
d1=pd.DataFrame()
d1["Name"]=["shiv","ram","raj"]
d1["age"]=[23,35,35]
# in the age col > 25 i want to replace with 300
con=d1["age"]>25
# when condition is True : 300
# False : keep it as it is
d1['age']=np.where(con,300,d1["age"])
d1
```
