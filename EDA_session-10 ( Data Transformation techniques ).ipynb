# Data Transformation techniques

- Generally used for to convert normal distribution

- Because all statitical math analysis by assumption data follows Normal distribution 

- It is also avoid skew ness also

- We have some important transformations

   - Log transformation  
    
   - Exponential transformation 

   - Reciprocal transformation 
    
   - Square root transformation 

   - Power transformation 

**step-1: Read the required packages** 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**step-2 : Read the data**

```python
exp_data=np.random.exponential(size=1000)
exp_data[:10]

# We are consider exponential data set with random 1k observations
# using numpy package
```

```python
plt.hist(exp_data,bins=30)
```

```python
plt.hist(exp_data,bins=30)
plt.title("Exponential data")
plt.show()
```

```python
plt.hist(exp_data,bins=30,label="Exponential")
plt.legend()

plt.show()
```

**Normal distribution data ( norm_data )**

```python
normal_data=np.random.normal(size=1000)
plt.hist(normal_data)
```

**Step-3 : Log Transformation**

```python
- np.log is used for log transformation 

- generally log transformation will not convert data into normal

- it avoids skewness

- np.log means natural logorithm base= e
```

```python
log_data=np.log(exp_data)
log_data[:10]
```

```python
exp_data[:10]
```

```python
plt.hist(log_data,bins=30,label="Log Data")
plt.legend()
```

```python
plt.subplot(1,2,1).hist(exp_data,bins=40,label="Exponential data")
plt.legend()
plt.title("Original data")
plt.subplot(1,2,2).hist(log_data,bins=40,label="Log Transformation")
plt.legend()
plt.title("Original - Normal Data")
plt.show()
```

**Step-4 : Reciprocal Transformation**

```python
- Reciprocal transformation 
```

```python
# Take the exponential data
# rec_data = 1 / exp_data
# plot it 
# if the observations in exponential becoming to 0 
# at that time 1 / 0 = not defined so it looks like 0 in  the reciprocal 
rec_data= 1/exp_data
rec_data[:10]
```

```python
plt.hist(rec_data)
```

```python
reciprocal_data=np.reciprocal(exp_data)
reciprocal_data[:10]
plt.hist(reciprocal_data)
```
