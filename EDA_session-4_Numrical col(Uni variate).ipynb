# Percentile-Quantile #

- using numpy
- percentile : if we want 25p give 25  ===== range 1-100
- Quantile : if we want q=25p give 0.25  q=50p (0.50)  q=75p (0.75) ====== range between 0-1

**25p**

```python
np.percentile(p_wage,25)
```

```python
np.quantile(p_wage,0.25)
```
